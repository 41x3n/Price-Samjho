```
#Removing unnecessary column
df_Kolkata = df_Kolkata.drop(['Unnamed: 0'], axis=1)
```

```
#Cleaning availability_date column and converting it to datetime
df_Kolkata.loc[df_Kolkata['availability_date'] == '0', 'availability_date'] = None
df_Kolkata['availability_date'] = pd.to_datetime(df_Kolkata['availability_date'])
```

```
#Converting NaN value to 1 as most apartment has atleast 1 balcony
df_Kolkata['balcony'].unique()
df_Kolkata['balcony'].value_counts()
df_Kolkata['balcony'] = df_Kolkata['balcony'].fillna(1.0)
df_Kolkata['balcony'] = df_Kolkata['balcony'].astype(int)
```
